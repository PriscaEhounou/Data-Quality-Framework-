```python
import os
import pandas as pd
```

```python
# 1.Load data
df = pd.read_csv("mortgage_data.csv")

df
```

```python
# 2.Check missing values


print("\n---Count of missing values by column ---")
missing_values = df.isnull().sum()

missing_values
```

```python
# 3. Check duplicate rows

print("\n---Return duplicate rows ---")
duplicate_rows = df[df["Loan_ID"].duplicated(keep=False)]

duplicate_rows
```

```python
# 4.Check invalid loan amount

print("\n---Rows where Loan_Amount is missing or not greater than 0 ---")
invalid_loan_amount = df[(df["Loan_Amount"].isnull()) | (df["Loan_Amount"] <= 0)]

invalid_loan_amount
```

```python
# 5.Check Invalid Interest Rate

print("\n---Rows where Interest_Rate is missing or outside 0 to 15---")
invalid_interest_rate = df[
        (df["Interest_Rate"].isnull()) |
        (df["Interest_Rate"] < 0) |
        (df["Interest_Rate"] > 15)
    ]

invalid_interest_rate
```

```python
# 6. Check missing or blank borrower names
print("\n---Rows where Borrower_Name is missing or blank---")
missing_borrower_name = df[(df["Borrower_Name"].isnull()) | (df["Borrower_Name"].astype(str).str.strip() == "")]

missing_borrower_name
```

```python
# 7. Check Invalid Loan status

print("\n---Rows where Loan_Status is not in allowed values---")
valid_statuses = ["Approved", "Pending", "Rejected"]
invalid_loan_status = df[~df["Loan_Status"].isin(valid_statuses)]


    
```

```python
# 8. Calculate total issues

total_issues = (
    len(duplicate_rows) +
    len(invalid_loan_amount) + 
    len(invalid_interest_rate) + 
    len(missing_borrower_name) + 
    len(invalid_loan_status) +
    missing_values.sum()
)

total_issues
```
